##Step 6: Data Visualization

```python
# Assigning data to df variable
df = pd.DataFrame(data)

# Creating a pie chart for Educational Qualifications
df['Educational Qualifications'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Educational Qualifications',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating pie chart for Monthly Income
colors = ['gold', 'yellowgreen', 'lightcoral']
df['Monthly Income'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=colors,startangle=90)
plt.title('Monthly Income',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))
plt.xticks(rotation=140)
plt.show()
```

```python
# Creating bar chart for top 10 Feedback
plt.figure(figsize=(6,4))
df['Feedback'].value_counts().head(10).plot(kind='bar',color='#FF0000',fontsize=10,edgecolor='black',width=0.8)
plt.xlabel('Feedback',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.title('Top 10 Feedback',weight='bold',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating  bar chart of top 10 Family size
plt.figure(figsize=(6,4))
df['Family size'].value_counts().head(10).plot(kind='bar',color='#1f77b4',edgecolor='black',width=0.8)
plt.xlabel('Family size')
plt.ylabel('Count')
plt.title('Top 10 Family size')
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating histogram of Monthly Income distribution
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.hist(df['Monthly Income'],color='#ff7f0e',edgecolor='black')
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')

# Creating histogram of Occupation distribution
plt.subplot(2, 2, 2)
plt.hist(df['Occupation'], color='#ff7f0e',edgecolor='black')
plt.title('Occupation Distribution')
plt.xlabel('Occupation')
plt.ylabel('Frequency')

# Creating histogram of Marital Status distribution
plt.subplot(2, 2, 3)
plt.hist(df['Marital Status'],color='#ff7f0e', edgecolor='black')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status (single/married)')
plt.ylabel('Frequency')

# Creating histogram of Age distribution
plt.subplot(2, 2, 4)
plt.hist(df['Age'], color='#ff7f0e',edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Creating bar chart of latitude in a particular Age
plt.figure(figsize=(8,4))
df['Age'].value_counts().sort_index().plot(kind='bar',color='#2ca02c',edgecolor='black',width=0.8)
plt.title('Count of latitude by Age',weight='bold',fontsize=14)
plt.xlabel('Age',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
```

```python
# Creating line chart to know trend of selling price over years
plt.figure(figsize=(8,4))
df.groupby('Gender')['Age'].mean().plot(marker='o')
plt.xlabel('Gender',weight='bold',fontsize=14)
plt.ylabel('Average Age',weight='bold',fontsize=14)
plt.title('Trend of Gender over Age',weight='bold',fontsize=14)
plt.show()
```
